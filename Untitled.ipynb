```python
import astropy.units as u
import astropy.constants as c
import numpy as np
import matplotlib.pyplot as plt
from astroquery.gaia import Gaia
```

```python
# Define and execute the ADQL query to crossmatch with Gaia DR3
query = f"""
SELECT *,
DISTANCE(POINT(132.825, 11.8), POINT(ra, dec)) AS ang_sep
FROM gaiadr3.gaia_source as gaia
WHERE DISTANCE(132.825, 11.8, ra, dec) < 1.
AND phot_g_mean_mag < 14
"""

# Upload the source_id table for crossmatching
job = Gaia.launch_job_async(query=query)
gaiadr = job.get_results()
```
